```python
import os
import zipfile
from google.colab import files
# create a zip archive of the directory
def zipdir(path, ziph):
    for root, dirs, files in os.walk(path):
        for file in files:
            ziph.write(os.path.join(root, file))

zipf = zipfile.ZipFile('/content/ARF-svox2/opt/ckpt_arf/llff/3Dart_result.zip', 'w', zipfile.ZIP_DEFLATED)
zipdir('/content/ARF-svox2/opt/ckpt_arf/llff', zipf)
zipf.close()

# download the zip archive
files.download('/content/ARF-svox2/opt/ckpt_arf/llff/3Dart_result.zip')
```
